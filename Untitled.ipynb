## Grabbing All the Elements of a Class
** We use wikipedia page of Grace Hopper

```python
#Grab the content of table ,we want to inspect a particular elements on a page and grab associated class
# create a new request
res = requests.get('https://en.wikipedia.org/wiki/Grace_Hopper')
```

```python
soup = bs4.BeautifulSoup(res.text,"lxml")
```

```python
#Here we grab the entire content of web page
#soup
```

```python
#select the particular class 
soup.select('.vector-toc-text')
```

```python
type(soup.select('.vector-toc-text')[0])
```

```python
#suppose we want particular text of a class
first_item = soup.select('.vector-toc-text')[7]
```

```python
#the text attribute returns the avaibale text between the spang
first_item.text
```

```python
#we want to print all those strings ,I could write the folling code
for item in soup.select('.vector-toc-text'):
    print(item.text)
```

```python
#Now we save it on our computer to do this we open a file 
#choose the here the name is my_computer_jpg,here i use wb for write binary
f = open('my_computer.jpg','wb')
```

```python
f.write(image_link.content)
```

```python
#Now we close it and see the image file locally
f.close()
```
